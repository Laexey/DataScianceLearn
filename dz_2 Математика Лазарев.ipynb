Загрузим датасет о ценах на дома в Бостоне. 

```python
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.data.shape)
```

```python
import pandas as pd
import numpy as np

data = pd.DataFrame(boston.data)
data.columns = boston.feature_names
```

Каждый дом - это вектор в 13-мерном пространстве. 
Написать код для подсчета скалярного произведения двух первых домов и датасета.

```python
x1 = boston.data[0] # дом 1
x2 = boston.data[1] # дом 2
```

```python
x1.dot(x2)
```

Нарисовать два дома в 3-мерном пространстве (атрибуты выбрать самостоятельно)

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline

figure = plt.figure()
ax = Axes3D(figure)
ax.plot([0,data.RM.loc[0]], [0,data.LSTAT.loc[0]], [0,data.AGE.loc[0]])
ax.plot([0,data.RM.loc[1]], [0,data.LSTAT.loc[1]], [0,data.AGE.loc[1]])

ax.scatter(data.RM.loc[0], data.LSTAT.loc[0], data.AGE.loc[0], marker='>')
ax.scatter(data.RM.loc[1], data.LSTAT.loc[1], data.AGE.loc[1], marker='>')

ax.set_xlabel("RM")
ax.set_ylabel("LSTAT")
ax.set_zlabel("AGE")
```

Найти косинус угла между векторами x1 и x2

```python
from numpy import linalg
import math

x1 = np.array([data.RM.loc[0], data.LSTAT.loc[0], data.AGE.loc[0]])
x2 = np.array([data.RM.loc[1], data.LSTAT.loc[1], data.AGE.loc[1]])

x1_norm = math.sqrt(x1[0]**2+x1[1]**2+x1[2]**2)
x2_norm = math.sqrt(x2[0]**2+x2[1]**2+x2[2]**2)

x1.dot(x2)/(x1_norm*x2_norm)
```

Найти расстояние между точками x1 и x2.

```python
math.sqrt((x1[0]-x2[0])**2+(x1[1]-x2[1])**2+(x1[2]-x2[2])**2)
```

# Задание повышенного уровня.

Задан объект в пространстве 2D. 
Изменить положение объекта (сместить по оси x на 300 пикселей, по оси y на 200 пикселей).

```python
import numpy as np
import matplotlib.pyplot as plt

A = [[-100, -90, 1],
    [-100, 90, 1],
    [100, 90, 1],
    [0, 0, 1],
    [100, -90, 1],
    [-100, -90, 1]]
A = np.array(A)

x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
x = A[:,0]
y = A[:,1]
plt.plot(x, y)

B = A + [300,200, 1]
x = B[:,0]
y = B[:,1]
plt.plot(x, y)
plt.ylim([-200, 500])
plt.xlim([-200, 500])
plt.legend(['Исходный','Смещенный'])
```

Уменьшить объект в два раза.

```python
A = np.array(A)

x = A[:,0]
y = A[:,1]

plt.plot(x, y)

C = A/2-[25,0,0]

plt.plot(C[:,0], C[:,1])

plt.ylim([-200, 200])
plt.xlim([-200, 200])

plt.legend(['Исходный','Уменьшенный'])
```

Повернуть на 130 градусов.

```python
from math import cos, sin, pi

theta = 130 * pi/180
TransformArray = np.array([
    [cos(theta), sin(theta), 0],
    [-sin(theta), cos(theta), 0],
    [0, 0, 1]])
```

```python
B = []
for point in A:
    new_point=np.dot(TransformArray,point)
    B.append(new_point)
B = np.array(B).astype(int)
print(B)
```

```python
plt.figure(figsize=(5,5))
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
x = B[:,0]
y = B[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.legend(['Исходный', 'Новый'])
```

Отразить объект относительно прямой y=x

```python

B = A + [100,0,0]
x = B[:,0]
y = B[:,1]

plt.plot(x, y)

C = B*-1

plt.plot(C[:,0], C[:,1])

plt.ylim([-200, 200])
plt.xlim([-200, 200])

plt.legend(['Исходный','Отраженный'])
```
