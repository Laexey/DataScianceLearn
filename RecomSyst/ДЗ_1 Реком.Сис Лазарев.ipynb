```python
import pandas as pd
```

```python
movies = pd.read_csv("movies.csv")
links = pd.read_csv('links.csv')
ratings = pd.read_csv('ratings.csv')
tags = pd.read_csv('tags.csv')
```

```python
movies_with_ratings = movies.join(ratings.set_index('movieId'), on='movieId')
```

```python
movies_with_ratings.groupby('title')[['rating']].mean().sort_values(by='rating', ascending=False)[:20]
```

```python
tagsqtt = tags.groupby('movieId')[['tag']].count()+1
```

```python
movies_with_tags = movies_with_ratings.join(tagsqtt, on='movieId')
```

```python
movies_with_tags.head()
```

```python
movies_with_tags.groupby('title')[['tag']].count().sort_values(by='tag', ascending=False)[:20]
```

### ДЗ: Постройте топ фильмов в категориях Action и Comedy (или любых других).

**Для начала можно строку с перечнем жанров разбить в список**

```python
movies_with_tags['genres'] = movies_with_tags.apply(lambda x: x['genres'].split('|'), axis=1)
```

```python
def func(x, y):
    for item in x:
        if item==y:
            return y
    return x
```

```python
movies_comedy=movies_with_tags.copy()
movies_action=movies_with_tags.copy()
```

**Теперь из списка жанров оставляем только те, что нас интересуют**

```python
movies_comedy['genres'] = movies_with_tags['genres'].apply(func, y='Comedy')
movies_action['genres'] = movies_with_tags['genres'].apply(func, y='Action')
```

**Выводим первые 20 по количеству упоминаний**

```python
movies_comedy[movies_comedy.genres=='Comedy'].groupby('title')[['tag']].count().sort_values(by='tag', ascending=False)[:20]
```

```python
movies_action[movies_action.genres=='Action'].groupby('title')[['tag']].count().sort_values(by='tag', ascending=False)[:20]
```
